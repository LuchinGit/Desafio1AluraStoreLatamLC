# 2. Ventas por categoría

```python
#Manera Básica
print("Tienda 1")
print(tienda['Categoría del Producto'].value_counts())

print("\nTienda 2")
print(tienda2['Categoría del Producto'].value_counts())

print("\nTienda 3")
print(tienda3['Categoría del Producto'].value_counts())

print("\nTienda 4")
print(tienda4['Categoría del Producto'].value_counts())
```

```python
# Agrupar las ventas por categoría para cada tienda
ventas_tienda1 = tienda.groupby('Categoría del Producto')['Precio'].sum().reset_index()
ventas_tienda2 = tienda2.groupby('Categoría del Producto')['Precio'].sum().reset_index()
ventas_tienda3 = tienda3.groupby('Categoría del Producto')['Precio'].sum().reset_index()
ventas_tienda4 = tienda4.groupby('Categoría del Producto')['Precio'].sum().reset_index()

# Mostrar los resultados de cada tienda
print("Ventas Tienda 1 por Categoría:")
print(ventas_tienda1)
print("\nVentas Tienda 2 por Categoría:")
print(ventas_tienda2)
print("\nVentas Tienda 3 por Categoría:")
print(ventas_tienda3)
print("\nVentas Tienda 4 por Categoría:")
print(ventas_tienda4)

# **Gráfica de ventas por categoría**
# Vamos a graficar las ventas por categoría para las 4 tiendas
# Para hacer las gráficas más comprensibles, vamos a graficarlas una por una.

def graficar_ventas_por_categoria(ventas_tienda, tienda_nombre):
    plt.figure(figsize=(10, 6))
    plt.bar(ventas_tienda['Categoría del Producto'], ventas_tienda['Precio'], color='green')
    plt.title(f'Ventas por Categoría - {tienda_nombre}', fontsize=16)
    plt.xlabel('Categoría del Producto', fontsize=12)
    plt.ylabel('Ventas Totales', fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

# Graficar para cada tienda
graficar_ventas_por_categoria(ventas_tienda1, 'Tienda 1')
graficar_ventas_por_categoria(ventas_tienda2, 'Tienda 2')
graficar_ventas_por_categoria(ventas_tienda3, 'Tienda 3')
graficar_ventas_por_categoria(ventas_tienda4, 'Tienda 4')
```

# 5. Envío promedio por tienda

```python
envio_prom_tienda1 = round(tienda["Costo de envío"].mean())
envio_prom_tienda2 = round(tienda2["Costo de envío"].mean())
envio_prom_tienda3 = round(tienda3["Costo de envío"].mean())
envio_prom_tienda4 = round(tienda4["Costo de envío"].mean())

print(f"Costo de envío promedio - Tienda 1: {envio_prom_tienda1} CLP")
print(f"Costo de envío promedio - Tienda 2: {envio_prom_tienda2} CLP")
print(f"Costo de envío promedio - Tienda 3: {envio_prom_tienda3} CLP")
print(f"Costo de envío promedio - Tienda 4: {envio_prom_tienda4} CLP")
```

```python
costos_envio = [envio_prom_tienda1, envio_prom_tienda2, envio_prom_tienda3, envio_prom_tienda4]
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

plt.stackplot(tiendas, costos_envio, color="blue")
plt.title("Costo de envío promedio por tienda (CLP)")
plt.ylabel("Pesos Chilenos (CLP)")
plt.xlabel("Tiendas")
plt.ylim(0, max(costos_envio) + 500)
plt.tight_layout()
plt.show()
```

**Análsis de cada tienda**

```python
def analizar_tienda(df, nombre):
    ingresos_totales = df["Precio"].sum()
    promedio_calificacion = df["Calificación"].mean()
    costo_envio_promedio = df["Costo de envío"].mean()
    ventas_por_categoria = df["Categoría del Producto"].value_counts()

    print(f"🔹 {nombre}")
    print(f"Ingresos totales: ${ingresos_totales:,.2f} CLP")
    print(f"Valoración media: {promedio_calificacion:.2f}")
    print(f"Costo de envío promedio: ${costo_envio_promedio:.2f} CLP")
    print("Ventas por categoría:")
    print(ventas_por_categoria)
    print("-" * 40)

# Ejecutar para cada tienda
analizar_tienda(tienda, "Tienda 1")
analizar_tienda(tienda2, "Tienda 2")
analizar_tienda(tienda3, "Tienda 3")
analizar_tienda(tienda4, "Tienda 4")
```
